```python
#Reading total sale data for last year for each product using a Pie chart with coding
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("C:\\Users\\nsr18\\OneDrive\\Desktop\\Mysql syllabus\\company_sales_data.csv")
print(df)

x=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
y=[df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
plt.pie(y,labels=x,autopct='%1.1f%%')
plt.show()
```
